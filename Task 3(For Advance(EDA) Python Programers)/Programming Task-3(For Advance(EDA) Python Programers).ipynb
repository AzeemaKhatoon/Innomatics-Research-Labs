### Q1. Maximum Product of Two Elements in an Array
-  Return the maximum value of (nums[i]-1)*(nums[j]-1).<br>
Example:

- Input: nums = [3,4,5,2]
- Output: 12 
- Explanation: If you choose the indices i=1 and j=2 (indexed from 0), you will get the maximum value, that is, (nums[1]-1)*(nums[2]-1) = (4-1)*(5-1) = 3*4 = 12.  

```python
def maxproduct(nums):
    
    nums.sort(reverse=False)
    max_prod = (nums[-1]-1)*(nums[-2]-1) 
            
    return max_prod
```

```python
maxproduct([3,4,5,2])
```

### Q2. Count Number of Teams
- Choose 3 soldiers with index (i, j, k) with rating (rating[i], rating[j], rating[k]).
- A team is valid if: (rating[i] < rating[j] < rating[k]) or (rating[i] > rating[j] > rating[k]) where (0 <= i < j < k < n).
- Return the number of teams you can form given the conditions. (soldiers can be part of multiple teams).<br>
Example:

- Input: rating = [2,5,3,4,1]
- Output: 3
- Explanation: We can form three teams given the conditions. (2,3,4), (5,4,1), (5,3,1). 

```python
def num_teams(rating):
    teams = 0
    for i in range(len(rating)):
        for j in range(len(rating)):
            for k in range(len(rating)):
                if ((rating[i] < rating[j] < rating[k]) or (rating[i] > rating[j] > rating[k])) and (0 <= i < j < k < len(rating)):
                    teams += 1
    return teams
```

```python
num_teams([2,5,3,4,1])
```

### Q3. Number of Students Doing Homework at a Given Time
- The ith student started doing their homework at the time startTime[i] and finished it at time endTime[i].
- Return the number of students doing their homework at time queryTime.<br>
Example:

- Input: startTime = [1,2,3], endTime = [3,2,7], queryTime = 4
- Output: 1
- Explanation: We have 3 students where:
The first student started doing homework at time 1 and finished at time 3 and wasn't doing anything at time 4.
The second student started doing homework at time 2 and finished at time 2 and also wasn't doing anything at time 4.
The third student started doing homework at time 3 and finished at time 7 and was the only student doing homework at time 4.

```python
def On_time(startTime, endTime, queryTime):
    
    student = 0
    for i in range(len(startTime)):
        if startTime[i] <= queryTime and endTime[i] >= queryTime:
            student += 1
    return student
```

```python
On_time(startTime=[1,2,3], endTime=[3,2,7], queryTime=4)
```

### Q4. Number of Steps to Reduce a Number to Zero
- In one step, if the current number is even, you have to divide it by 2, otherwise, you have to subtract 1 from it.<br>
Example:

- Input: num = 14
- Output: 6
- Explanation: <br>
Step 1) 14 is even; divide by 2 and obtain 7. <br>
Step 2) 7 is odd; subtract 1 and obtain 6.<br>
Step 3) 6 is even; divide by 2 and obtain 3. <br>
Step 4) 3 is odd; subtract 1 and obtain 2. <br>
Step 5) 2 is even; divide by 2 and obtain 1. <br>
Step 6) 1 is odd; subtract 1 and obtain 0.

```python
def no_of_step(num):
    n = num
    step = 0
    for i in range(n):
    
        if num%2 == 0:
            div = num//2
            if div%2 != 0:
                sub = div - 1 
                if sub == 0:
                    step += 2
                    break
                else:
                    num = sub
                    step += 2
                    if num == 0:
                        break
            else:
                 num = div//2 
                 step += 2
                 if num == 0:
                     break
        else:
            num = num - 1 
            step += 1
            if num == 0:
                break
    return step
```

```python
no_of_step(14)
```

### Q5. Counting Bits
- binary representation <br>
Example:

- Input: n = 2
- Output: [0,1,1]
- Explanation: <br>
0 --> 0 <br>
1 --> 1 <br>
2 --> 10

```python
import re
def bitslist(num):
    li = []
    li1 = []
    li2 = []
    for i in range(num+1):
        li.append(bin(i))
        s = re.sub(r"[0b]", '', li[i])
        if s == "":
            li1.append("0")
        else:
            li1.append(s)
        count = 0
        for k in li1[i]:
            if k == "0":
                count = 0
            else:
                count += 1
        li2.append(count)
    return li2
```

```python
bitslist(2)
```
