### Q1. Defanging an IP Address
- A defanged IP address replaces every period "." with "[.]".<br>
 Example:
- Input: address = "1.1.1.1"
- Output: "1[.]1[.]1[.]1"

```python
import re

def ip_address(address):

    adr = re.sub(r"[/.]", "[.]", address)
    return adr
```

```python
ip_address("1.1.1.1")
```

### Q3. Number of Good Pairs
- A pair (i, j) is called good if nums[i] == nums[j] and i < j.<br>
Example:

- Input: nums = [1,2,3,1,1,3]
- Output: 4
- Explanation: There are 4 good pairs (0,3), (0,4), (3,4), (2,5) 0-indexed.

```python
def goodpairs(nums):
    
    count = 0
    for i in range(len(nums)):
        for j in range(len(nums)):
    
            if nums[i] == nums[j] and i < j:
                count += 1
    return count
```

```python
goodpairs([1,2,3,1,1,3])
```

### Q4. How Many Numbers Are Smaller Than the Current Number
- find out how many numbers in the array are smaller than it.<br>
  Example:
- Input: nums = [8,1,2,2,3]
- Output: [4,0,1,1,3]
- Explanation: 
- For nums[0]=8 there exist four smaller numbers than it (1, 2, 2 and 3). 
- For nums[1]=1 does not exist any smaller number than it.
- For nums[2]=2 there exist one smaller number than it (1). 
- For nums[3]=2 there exist one smaller number than it (1). 
- For nums[4]=3 there exist three smaller numbers than it (1, 2 and 2).

```python
import numpy as np
def smaller_num_than_current(nums):
    
    li = [] 
    for i in range(len(nums)):
        count = 0
        for j in range(len(nums)):
    
            if nums[j] < nums[i]: 
                count += 1
        li.append(count)
    return li
```

```python
smaller_num_than_current([8,1,2,2,3])
```

### Q5. Subtract the Product and Sum of Digits of an Integer
- number n, return the difference between the product of its digits and the sum of its digits.<br>
 Example:

- Input: n = 234
- Output: 15 
- Explanation: 
- Product of digits = 2 * 3 * 4 = 24 
- Sum of digits = 2 + 3 + 4 = 9 
- Result = 24 - 9 = 15

```python
def diff_of_prod_sum(nums):
    
    prod = 1
    add = 0
    for i in str(nums):
        prod *= int(i)
        add += int(i)
        
    return prod - add 
```

```python
diff_of_prod_sum(234)
```

### Q6. XOR Operation in an Array
- nteger n and an integer start.

- Define an array nums where nums[i] = start + 2 * i (0-indexed) and n == nums.length.

- Return the bitwise XOR of all elements of nums.<br>

  Example:

- Input: n = 5, start = 0
- Output: 8
- Explanation: Array nums is equal to [0, 2, 4, 6, 8] where (0 ^ 2 ^ 4 ^ 6 ^ 8) = 8.
Where "^" corresponds to bitwise XOR operator.

```python
def xoroperation(n, start):

    s = 0
    for i in range(n):
        o = start+2*i
        s = s^o
    return s
```

```python
xoroperation(n=5, start=0)
```
